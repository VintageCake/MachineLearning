```python
# K Nearest Neighbors for a line (i think)
# ls223qx@student.lnu.se Love Samuelsson

import numpy as np
import math
from numpy import genfromtxt
from collections import Counter
import matplotlib.pyplot as plt

data = genfromtxt('polynomial200.csv',delimiter=',') # Imports data


## Splitting data into training/test features and targets
X_train = data[0:100:,0]
Y_train = data[0:100:,1]
X = data[100:200:,0]
Y = data[100:200:,1]

## Creating 200 equadistant points on x axis based on training data
grid_size = 200
min_x, max_x = min(X_train), max(X_train)
x_axis = np.linspace(min_x, max_x, grid_size)
```

```python
def _2dDistFunc(p0, p1):
    return math.hypot( p1[0]-p0[0], p1[1] - p0[1] )

def _1dDistFunc(x0,x1):
    return abs(x0-x1)

## Return single average y value to a corresponding x
# Inputs are the current x and y sets, the point to be tested and the k value
def predict_line(x,y,linspaceValues,k):
    y_grid = []
    for linspaceX in linspaceValues:
        distances =  [_1dDistFunc(xpoint, linspaceX) for xpoint in x]
        indecies = np.argsort(distances)
        
        closestValues = []
        for index in indecies[:k]:
            closestValues.append(y[index])
        
        sums = np.sum(closestValues)
        average = sums / k
        y_grid.append(average)
    return y_grid

def predict_single(x,y,testPoint,k):
    distances =  [_1dDistFunc(xpoint, testPoint) for xpoint in x]
    indecies = np.argsort(distances)
        
    closestValues = []
    for index in indecies[:k]:
        closestValues.append(y[index])
        
    sums = np.sum(closestValues)
    return sums / k
        
    
        
```

```python
fig, axs = plt.subplots(3,2, figsize=(15,15))
axs = axs.flatten()

kval=[1,3,5,7,9,11]
## Lines contains the trained 'model'
lines = []
for k in kval:
    lines.append(predict_line(X_train,Y_train,x_axis,k))

for index,ax in enumerate(axs):
    ax.plot(x_axis,lines[index])
    ax.scatter(X_train, Y_train, s=3)
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    
    predicts = []
    for x in X_train:
        predicts.append(predict_single(X_train,Y_train,x,kval[index]))
    
    # Taken from https://www.geeksforgeeks.org/python-mean-squared-error/
    # Calculates the mean-square error by performing an equation of all predicted Y values and all true Y values.
    MSE = np.square(np.subtract(Y_train,predicts)).mean()
    ax.set_title(f"Training data with k = {kval[index]} and MSE = {MSE}")

plt.show()
```

```python
fig, axs = plt.subplots(3,2, figsize=(15,15))
axs = axs.flatten()

## Iterates over all plots
for index,ax in enumerate(axs):
    ax.plot(x_axis,lines[index]) # Plots the trained model line from the training data
    ax.scatter(X, Y, s=3) # Scatters the regular data
    ax.set_xlabel("x")
    ax.set_ylabel("y")
    
    predicts = []
    for x in X: # Performs predictions of current true test data compared to training data
        predicts.append(predict_single(X_train,Y_train,x,kval[index]))
    
    # Taken from https://www.geeksforgeeks.org/python-mean-squared-error/
    # Calculates the mean-square error by performing an equation of all predicted Y values and all true Y values.
    MSE = np.square(np.subtract(Y,predicts)).mean()
    ax.set_title(f"Test data with k = {kval[index]} and MSE = {MSE}")

plt.show()
```
