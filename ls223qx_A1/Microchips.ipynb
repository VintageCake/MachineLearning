```python
# K Nearest Neighbors for a bunch of microchips
# ls223qx@student.lnu.se Love Samuelsson

import numpy as np
import math
from numpy import genfromtxt
from collections import Counter
import matplotlib.pyplot as plt

data = genfromtxt('microchips.csv',delimiter=',') # Imports data

X_train = data[:,:2]
Y_train = data[:,2]
```

```python
### Get distance between two points
def distFunc(chip0, chip1):
    return math.hypot( chip1[0]-chip0[0], chip1[1] - chip0[1] )

### KNN Classifier
def predict(X_train, Y_train, point, k):
    dist = [distFunc(x, point) for x in X_train] # funky syntax creates list
    #Sort by nearest
    indecies = np.argsort(dist)
    x_point = X_train[indecies[0]]
    
    checkedValues = []
    for index in indecies[:k]:
        checkedValues.append(Y_train[index])
    
    mostCommonValue = Counter(checkedValues).most_common(1)
    # Returns either 1.0 or 0.0 depending on which appeared most often
    return mostCommonValue[0][0]
    
```

```python
### Plotting! Equivalent to task 1

twoArrays = [data[data[:,2]==k] for k in np.unique(data[:,2])] # Splits into two np arrays, one with all 0 results and one with all 1s

falseData = twoArrays[0]
trueData = twoArrays[1]

plt.figure(figsize=(15,5))
plt.subplot(132)
plt.scatter(falseData[:,0], falseData[:,1], label='OK')
plt.scatter(trueData[:,0], trueData[:,1], label="Fail")
plt.legend()


#plt.subplot(133)
#plt.scatter(X_train[:,0], X_train[:,1], c=Y_train)
#plt.legend()

#plt.subplot(133)
#plt.scatter(data[:,0],data[:,1], label="Microchip")
#plt.legend()


plt.show()
```

```python
### 2 - Does the kNN classification with k= 1,3,5,7 for each of the chips.
testChips = np.array( [[-0.3,1.0] , [-0.5,-0.1] , [0.6,0.0]])

for k in [1,3,5,7]:
    print(f"With k = {k}")
    for index,chip in enumerate(testChips):
        print(f"Testing {chip} chip# {index}" )
        test = predict(X_train, Y_train, chip, k)
        if (test) == 1:
            print ("--> OK")
        else:
            print ("--> Fail")
    print()
```

```python

### Setup helper variables for grid generation
grid_size = 100
min_x, max_x = min(X_train[:, 0]), max(X_train[:, 0])
min_y, max_y = min(X_train[:, 1]), max(X_train[:, 1])
x_axis = np.linspace(min_x, max_x, grid_size) # Equadistant points
y_axis = np.linspace(min_y, max_y, grid_size)

### Code partially taken from "Example decision boundaries" and then modified until death
def make_grid_opt(x_axis, y_axis, k):
    xx, yy = np.meshgrid(x_axis, y_axis)
    cells = np.stack([xx.ravel(), yy.ravel()], axis=1)
    grid = numpy.array([predict(X_train, Y_train, c, k) for c in cells])
    return grid.reshape(grid_size, grid_size)
    
### Generates prediction grids
grids = []
for k in [1,3,5,7]:
    grids.append(make_grid_opt(x_axis, y_axis, k))


### Starts plotting stuff
fig, axs = plt.subplots (2, 2, figsize=(15,10)) ## Arbritrary size
axs = axs.flatten() # flattening this stupid 2d matrix into 1d array

### Plots all data and all decision boundaries generated
kval = [1,3,5,7]
for gridIndex,grid in enumerate(grids):
    axs[gridIndex].imshow(grid, origin='lower', extent=(min_x, max_x, min_y, max_y)) # Puts the grid image on the thing!
    axs[gridIndex].scatter(X_train[:,0], X_train[:,1], c=Y_train, edgecolors='k')
    trainingErrors = 0
    for pointIndex,x in enumerate(X_train): ## Performs the training error check
        if predict(X_train, Y_train, x, kval[gridIndex]) != Y_train[pointIndex]:
            trainingErrors += 1 ## if training data prediction is wrong, iterate
    axs[gridIndex].set_title(f"Decision Boundary k = {kval[index]} with training error = {trainingErrors}")

plt.show()
```
